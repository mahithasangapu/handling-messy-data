Task 1

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age': [25, np.nan, 35, np.nan, np.nan, 40],
    'city': ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})


df['age'].fillna(df['age'].median(), inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)


df = df.drop_duplicates()

print(df)
```

Task 2

```python

df = pd.get_dummies(df, columns=['city'])
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})

print(df)
```

task 3

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[['age', 'annual_income']] = minmax.fit_transform(df[['age', 'annual_income']])

print("Min-Max Scaling:\n", df_minmax)

standard = StandardScaler()
df_standard = df.copy()
df_standard[['age', 'annual_income']] = standard.fit_transform(df[['age', 'annual_income']])

print("\nStandard Scaling:\n", df_standard)
```
